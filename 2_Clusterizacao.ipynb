### Remoção de Outliers

```python
colunas_a_verificar = ['valor_total', 'numero_visitas', 'quantidade_produtos', 'dias_desde_ultima_compra']

for coluna in colunas_a_verificar:
    Q1 = df0[coluna].quantile(0.25)
    Q3 = df0[coluna].quantile(0.75)
    
    IQR = Q3 - Q1
    
    #Removendo os outliers
    df1 = df1.reset_index(drop=True)
    df1 = df1.loc[~((df1[coluna] < (Q1 - 1.5 * IQR)) | (df0[coluna] > (Q3 + 1.5 * IQR)))]
```

## Descobrindo o número ideal de clusters

```python
#Colunas relevantes para clusterização
X = df1[['valor_total', 'numero_visitas', 'quantidade_produtos', 'dias_desde_ultima_compra']]

#Normalizando os valores para estarem na mesma escala
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Instância do modelo
kmeans = KMeans()

#Criando uma visualiação do cotovelo usando o KElbowVisualizer
visualizador = KElbowVisualizer(kmeans, k=(2,10))
visualizador.fit(X_scaled)

#Mostrando a visualização
visualizador.show()
```

### Baseado no Método do Cotovelo, verificamos que o número ideal de clusters/grupos = 4

```python
kmeans_prod = KMeans(n_clusters=4, random_state=0, init='k-means++')
kmeans_prod.fit(X)
```

```python
# Adicionando as etiquetas de cluster ao DataFrame
df1['cluster'] = kmeans_prod.labels_
```

```python
#Imprimindo o centro dos clusters
print('Centro dos clusters: {}'.format(kmeans_prod.cluster_centers_))
```

## PCA - Análise de Componentes Principais

Técnica estatística que permite reduzir a dimensionalidade dos dados, mantendo a maior parte da variância dos dados originais.

Nesse caso, reduzimos das 4 dimensões(colunas) para 2 dimensões.

A análise de componentes principais pode ser útil para visualizar a estrutura dos dados em um espaço de dimensionalidade reduzida, o que pode ajudar a identificar padrões e estruturas nos dados que não seriam aparentes em um espaço de dimensionalidade mais alta. No contexto da clusterização, a PCA pode ser usada para visualizar as relações entre os diferentes clusters identificados pelo algoritmo, bem como para avaliar a qualidade da clusterização.

Para usar essa análise, você pode examinar o gráfico de dispersão criado pelo código para identificar padrões e estruturas nos dados. Por exemplo, você pode procurar clusters que são bem separados uns dos outros, o que pode indicar que a clusterização foi bem-sucedida. Você também pode procurar padrões nas componentes principais, como grupos de pontos que se aglomeram em determinada região do gráfico, o que pode indicar a presença de subestruturas nos dados.

Além disso, você pode usar a análise de componentes principais para avaliar a qualidade da clusterização. Por exemplo, se os clusters identificados pelo algoritmo K-Means são bem separados em relação às duas primeiras componentes principais, isso pode indicar que a clusterização foi bem-sucedida em capturar as principais estruturas dos dados. Se os clusters são sobrepostos ou misturados em relação às componentes principais, isso pode indicar que a clusterização não foi tão bem-sucedida em capturar as principais estruturas dos dados.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Definir o número de componentes para a PCA
pca_num_components = 2

# Aplicar a PCA aos dados do DataFrame df1
pca = PCA(n_components=pca_num_components)
reduced_data = pca.fit_transform(df1)

# Criar um DataFrame com os dados reduzidos
results = pd.DataFrame(reduced_data, columns=['PCA1', 'PCA2'])

# Criar um gráfico de dispersão usando seaborn
sns.scatterplot(x="PCA1", y="PCA2", hue=df1['cluster'], data=results)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```
